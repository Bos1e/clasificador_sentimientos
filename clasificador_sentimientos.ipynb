```python
# Paso 1: Instalar las dependencias
!pip install transformers torch

# Paso 2: Importar las librerías necesarias
from transformers import pipeline

# Paso 3: Crear el clasificador de sentimientos usando un modelo preentrenado
classifier = pipeline("sentiment-analysis")

# Paso 4: Clasificar un texto de ejemplo
text = "Me encanta este producto, es increíble!"
result = classifier(text)

# Paso 5: Mostrar el resultado
print(f"Sentimiento: {result[0]['label']}, Confianza: {result[0]['score']:.2f}")
```
